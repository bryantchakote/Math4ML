"Probabibilty distributions". What a strange but easy to understand concept with a proven importance in Data Science. By allowing us to understand the "shape" of the data at our disposal, they provide safeguards to pursue our analysis, orient the preprocessing and feature engineering phases, and help pre-selecting the more suitable models depending on the asumptions on the data. According to ChatGPT (don't worry, the rest of this article wasn't written by ChatGPT), a probability distribution refers to a mathematical function that describes the likelihood of different outcomes or events occurring in a particular situation or random experiment. It provides a way to assign probabilities to different possible values of a random variable.

A very intuitive way to think about this is to consider a random experiment, let's say tossing a coin. As a result, we either have **head** or **tail**. We're going to toss our coin several times, save the results progressively, and at the end draw a bar plot to summarize the number of occurences of each possible outcome. This is commonly known as the **probability mass function**, or the **probability density function** if the result of our experiment can take any value within an interval.

Throughout the ages (since 1713 with Bernouilli), people have conducted several random experiments, reported the results, and made particular remarks about some of them. The researchers could then derive interesting properties from particular random experiments, we're outlining some of the most popular ones.

```python
# Importing the required libraries
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')
import seaborn as sns

# Colors
red, green, blue, yellow = '#FF0060', '#00DFA2', '#0079FF', '#F6FA70'
```

# Discrete distributions

A random variable's probability distributions is qualified as discrete when its possible values are included in a countable set (finite or not).

## Bernouilli distribution

- Used to model the success in a single random experiment that results in exactly two outcomes: failure and success ($0$ and $1$).
- A single parameter, the success probability $p$
- Mean: $p$ - Variance: $p(1 - p)$

Consider an urn containing **4 red balls** and **1 blue ball**. Drawing a ball at random from this urn, calling it a **success if it is blue** and a **failure if it is red**, constitutes a Bernouilli test, with $p = \frac{1}{4+1} = \frac{1}{5} = 0.2$.

```python
# Generation of 10,000 samples from the Bernouilli distribution with p = 0.2
p, n_samples = .2, 10000
samples = np.random.binomial(1, p, n_samples)

# Mapping indexes to values
mapping = {0: 'Red', 1: 'Blue'}
samples = np.array([mapping.get(x, 'Unknown') for x in samples])

print('5 first balls drawed:', samples[:5])
```

```python
# Ploting our data
ax = sns.countplot(x=samples, color=red, alpha=.6, width=.4)

for bar in ax.patches:
    pct = bar.get_height() / n_samples # percentage of observations
    xy = bar.get_x() + bar.get_width() / 2, bar.get_height() / 2 # position
    ax.annotate(f'{pct:.2%}', xy, ha='center', va='center',
                size=15, color=yellow, fontweight=600)

plt.xlabel('Ball color')
plt.ylabel('Number of trials')
plt.title(f'Bernouilli distribution, p = {p}')
plt.show()
```

## Binomial distribution

- Defined as the sequence of $n$ independant Bernouilli trials. It focuses on the number $k$ of successes out of the $n$ trials
- Two parameters: number of trials $n$ and success probability $p$ to each trial
- Mean: $np$ - Variance: $np(1 - p)$

*A binomial distribution with parameters $(1, p)$ is equivalent to a Bernouilli with parameter $p$.*

Consider the first example, this time with one urn containing **3 red balls** and the second containing **7 blue balls, the targeted ones**. In the previous example, recording the color of one ball choosed at random constituted one trial of the random experiment. This time, the game is a little bit different: to achieve one trial, we need to go through 10 successive random draws **with replacement** and record the **total number of times that we got a blue ball**: 0, 2, 7, 10, ... This suggest a random variable with a binomial distribution having $n = 10$ and $p = \frac{7}{3+7} = \frac{7}{10} = 0.7$.

```python
# 10,000 samples of a binomial with parameters n = 10 and p = 0.7
n, p, n_samples = 10, .7, 10000
samples = np.random.binomial(n, p, n_samples)
print('Number of times a blue ball was picked (8 trials):', samples[:8])
```

```python
# Plotting
ax = sns.countplot(x=samples, color=yellow, alpha=.6)

E = round(n*p) # expectation / average number of successes

for i, label in enumerate(ax.get_xticklabels()):
    if int(label.get_text()) == E:
        ax.containers[0][i].set(color=green, label=f'Average : n x p = {E}')

plt.xlabel('Blue balls')
plt.ylabel('Number of trials')
plt.title(f'Binomial distribution, n = {n} and p = {p}')
plt.legend(loc='upper left')
plt.show()
```

## Multinomial distribution

- Generalisation of the binomial distribution: we transition from 2 possible outcomes to $m \in \mathbb{N}$, and we are interested on the number of occurences of each of the outcomes
- Have $m + 1$ parameters: the number of trials $n$ and the probabilities $p_{i=1, ..., m}$ of each possible outcomme. Obviously, $\sum_{i = 1}^m p_i = 1$

**North**, **East**, **South**, and **West** were the 4 presidential candidates in *World Country* in June 2023. We know that 5% of the eligible voters prefer **North**, 20% prefer **East**, 45% prefer **South**, and the remaining 30% prefer **West** (in real life we may also have abstentions). Recording the number of votes towards each candidate in a group of 10 eligible voters sampled randomly imply defining a multinomial random variable with $n = 10$ and $p_1 = 0.05$, $p_2 = 0.2$, $p_3 = 0.45$, and $p_4 = 0.3$. 

```python
# 10,000 samples of a multinomial distribution with parameters
# n = 10 and p = [0.05, 0.2, 0.45, 0.3]
n, p, n_samples = 10, [.05, .2, .45, .3], 10000
samples = np.random.multinomial(n, p, n_samples)
print('Number of voters for each candidate (5 trials):\n', samples[:5])
```

```python
# Extending `samples` to get the votes for each candidate
candidates = ['North', 'East', 'South', 'West']
x = []
for sample in samples:
    for candidate, i in zip(candidates, sample):
        x.extend([candidate]*i)
print('First 7 votes:', x[:7])

# Sort the votes in the order of the candidate names
x.sort(key=lambda y: candidates.index(y))
```

```python
ax = sns.countplot(x=x, color=blue, alpha=.6)

for bar in ax.patches:
    pct = bar.get_height() / len(x) # percentage of observations
    xy = bar.get_x() + bar.get_width() / 2, bar.get_height() / 2 # position
    ax.annotate(f'{pct:.2%}', xy, ha='center', va='center',
                size=15, color=yellow, fontweight=600)

plt.xticks(range(len(samples[0])))
plt.xlabel('Candidate')
plt.ylabel('Total number of votes')
plt.title(f'Multinomial distribution, n = {n} and p = {p}')
plt.legend()
plt.show()
```

## Poisson distribution

- Used to model the number of occurrences of rare and independent events within a given time window, or to study queuing phenomena
- Its parameter $\lambda$ represents both its mean and variance

An ATM in the capital of *World Country* receives everyday between 7 and 8 a.m. an average number of 3 visitors. Given a random morning, the number of visitors between 7 and 8 follows a Poisson distribution with $\lambda = 3$.

```python
# 10,000 samples of a Poisson Distribution with parameters λ = 3
l, n_samples = 3, 10000 # ~27 years of observations, sounds impossible
samples = np.random.poisson(l, n_samples)
print('Number of visitors during the first 5 days:', samples[:5])
```

```python
# Plotting
ax = sns.countplot(x=samples, color=green, alpha=.6)

for i, label in enumerate(ax.get_xticklabels()):
    if int(label.get_text()) == l:
        ax.containers[0][i].set(color=red, label=f'Occurences avg.: {l}')

plt.xlabel('Number of visitors at the ATM')
plt.ylabel('Number of observations')
plt.title(f'Poisson distribution, lambda = {l}')
plt.legend()
plt.show()
```

# Continuous distributions

We talk about a continuous distribution when a random variable can take any possible value within an intervall (bounded or not).

```python
def plot_distribution(distribution=None, params=[None, None],
                      n_samples=10000, bins=20, color=None):
    """
    This function generates `n_samples` observations from a `distribution`
    distribution with parameter(s) `params`, and plots on the same graph:
    - The histogram of the generated data with a number of bins equal to `bins`
    - The theoretical density function of the distribution
    The `color` parameter allows you to define the color of the plots
    """
    
    # Distributions list
    try:
        distributions = ['uniform', 'normal', 'chisquare',
                         'student', 'fisher', 'exponential']
        distribution = distribution.lower()
        assert distribution in distributions
    except AssertionError:
        print(f'`distribution` not in {distributions}.')
        return
    
    # `params` must be a list
    try:
        assert isinstance(params, list)
    except AssertionError:
        print(f'`params` should be a list.')
        return
    
    # The number of parameters should be consistent with the choosen distribution
    need_two_params = ['uniform', 'normal', 'fisher'] 
    if distribution in need_two_params: # distributions requiring 2 parameters
        try:
            assert len(params) == 2
            assert params[0] is not None and params[1] is not None 
        except AssertionError:
            print(f'The {distribution} distribution needs 2 parameters.')
            return
        p1, p2 = params
    else: # the others (only 1 required parameter)
        try:
            assert len(params) == 1
            assert params[0] is not None 
        except AssertionError:
            print(f'The {distribution} distribution needs 1 parameter.')
            return
        p = params[0]
    
    # Samples generation
    try:
        if distribution == 'uniform':
            samples = np.random.uniform(low=p1, high=p2, size=n_samples)
        elif distribution == 'normal':
            samples = np.random.normal(loc=p1, scale=p2, size=n_samples)
        elif distribution == 'chisquare':
            samples = np.random.chisquare(df=p, size=n_samples)
        elif distribution == 'student':
            samples = np.random.standard_t(df=p, size=n_samples)
        elif distribution == 'fisher':
            samples = np.random.f(dfnum=p1, dfden=p2, size=n_samples)
        elif distribution == 'exponential':
            samples = np.random.exponential(scale=1/p, size=n_samples)
    except ValueError as e:
        print('Error in `params`,', e)
        return
    print('Some values:', samples[:5])
    min_val, max_val = samples.min(), samples.max()
    print(f'Min: {min_val:.4f} - Max: {max_val:.4f}')
    
    # Plotting
    ## Empirical samples
    try:
        plt.hist(samples, bins=bins, range=(min_val, max_val),
                 density=True, color=color, alpha=.4)
    except ValueError as e:
        print('Error in `color`,', e)
        return
    
    ## Theoritical distribution
    x = np.linspace(min_val, max_val, 200)
    
    if distribution == 'uniform': y = uniform.pdf(x, loc=p1, scale=p2-p1)
    elif distribution == 'normal': y = norm.pdf(x, loc=p1, scale=p2)
    elif distribution == 'chisquare': y = chi2.pdf(x, df=p)
    elif distribution == 'student': y = t.pdf(x, df=p)
    elif distribution == 'fisher': y = f.pdf(x, dfn=p1, dfd=p2)
    elif distribution == 'exponential': y = expon.pdf(x, scale=1/p)
    
    try:
        plt.plot(x, y, label='Theoritical distribution', color=color)
    except ValueError as e:
        print('Error in `color`,', e)
        return
    
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.title(f'{distribution.title()} distribution, params: {params}')
    plt.legend()
    plt.tight_layout()
    return plt.gca()
```

## Uniform distribution

- Models an experiment where each value in the interval $[a, b]$ has an equal probability of being chosen
- Reference: $a = 0$ and $b = 1$ (uniform distribution on $[0, 1]$; $U_{[0, 1]}$)
- Mean: $\frac{a + b}{2}$ (center of the interval) - Variance: $\frac{(b - a)^2}{12}$
- Used for weight initialization in neural networks, optimal hyperparameter search, data augmentation, etc.

*A discrete probability distribution can also be referred to as uniform. In this case, the interval $[a, b]$ is replaced by a finite set with a cardinality of $n$, where each element has a probability of $\frac{1}{n}$ of being selected.*

A bus stop in *Unif City* has its 2 last buses everyday at 10:30 and 11 p.m. With the agreement of the town's residents, a well-tested, privacy-friendly AI system was installed to record passenger arrival times (and nothing else) over a given period. The histogram of the 10,000 observations collected suggests a uniform distribution of the variable of interest.

```python
from scipy.stats import uniform
ax = plot_distribution(distribution='uniform', params=[0, 0.5], # half an hour
                       n_samples=10000, bins=30, color=green)
ax.set_xlabel('Passengers arrival time')
ax.set_xticks(np.linspace(0, 0.5, 4), ['10:30', '10:40', '10:50', '11:00'])
plt.show()
```

## Normal (or gaussian) distribution

- Symmetric with respect to its mean and bell-shaped, its spread depends on its standard deviation
- Defined by two parameters: its mean $\mu$ and its standard deviation $\sigma$
- For $\mu = 0$ and $\sigma = 1$, we have the standard normal distribution: $X \sim N(0, 1)$
- Used to model many natural phenomena, weight initialization in neural networks, run many statistical tests, etc.

Fun fact: In *Normal City*, the company producing and commercializing natural mineral water never writes the exact contenance on its bottles. For example, the supposed **2L** bottles have this weird inscription: $volume \sim N(2, 0.05) \; liters$. *Ms Abnormal*, a citizen of that town implemented the means to order and measure 10,000 bottles, in order to figure out what was the meaning of this. Here are the results she obtained at the end of the day...

```python
from scipy.stats import norm
ax = plot_distribution(distribution='normal', params=[2, 0.05],
                       n_samples=10000, bins=40, color=red)
ax.set_xlabel('Bottle contenance')
plt.show()
```

## Chi-squared ($\chi^2$) distribution

- Positive and right-skewed.
- Defined by a single parameter, the degree of freedom (or df) $k$.
- Mean: $k$ - Variance: $2k$.
- Sum of squares of $k$ independent standard normal random variables: $Y = \sum_{i=1}^kX_i^2$.
- Used in statistical inference (test of independence between two categorical variables, test if a sample fits to a distribution, ANalysis Of VAriance (ANOVA)), etc.

We're going to generate samples from this distribution without prior context since it's often used in statistical inference.

```python
from scipy.stats import chi2
plot_distribution(distribution='chisquare', params=[5],
                  n_samples=10000, bins=40, color=blue)
plt.show()
```

## Student or t-distribution

- Symmetric, similar to the normal distribution but with longer tails
- One parameter, the degree of freedom $k$
- Mean: $0$ - Variance: $\frac{k}{k - 2}$.
- A standard normal random variable divided by the square root of a chi-squared with $k$ degrees of freedom over $k$: $Z = \frac{X}{\sqrt{Y / k}}$
- Used to construct confidence intervals for means, test if the means of two populations are significantly different, etc.

We're going to generate samples from this distribution without prior context since it's often used in statistical inference.

```python
from scipy.stats import t
plot_distribution(distribution='student', params=[6],
                  n_samples=10000, bins=40, color=red)
plt.show()
```

## Fisher of f-distribution

- Asymmetric
- Two parameters (degrees of freedom) $m$ and $n$
- Mean: $\frac{n}{n - 2}$ - Variance: kind of complicated formula... :-/
- Ratio of a chi-squared random variable $Y_1$ divided by its degrees of freedom $m$ to a chi-squared $Y_2$ divided by its degrees of freedom $n$: $F = \frac{Y_1 / m}{Y_2 / n}$
- Used to construct confidence intervals for variances, test if the variances of two populations are significantly different, in ANOVA, etc.

We're going to generate samples from this distribution without prior context since it's often used in statistical inference.

```python
from scipy.stats import f
plot_distribution(distribution='fisher', params=[6, 8],
                  n_samples=10000, bins=40, color=blue)
plt.show()
```

## Exponential distribution

- Very close to the Poisson distribution (which is discrete), it models the time gap between two events in a Poisson process
- It has only the average event rate $\lambda$ as a parameter
- Mean: $\frac{1}{\lambda}$ - Variance: $\frac{1}{\lambda^2}$

July is here! A lot of people have surely planned to move from *Unif City* to enjoy the sandy beaches of *Normal City*. There's a toll station on the highway separating the two towns which is said to receive a car every 4 seconds between 8 a.m. and 2 p.m during the summer. While you'll be travelling, *Mr Noholidays* will settle somewhere near the toll both (if it's authorized) to check whether the statement that the time between two cars follows an exponential distribution with a mean of 4 seconds is true.

```python
from scipy.stats import expon
ax = plot_distribution(distribution='exponential', params=[1/4],
                       n_samples=10000, bins=40, color=green)
ax.set_xlabel('Time elapsed between 2 cars')
plt.show()
```

As you have seen, probability distributions are very simple and intuitive. They help understanding how a phenomena is going on, and are widely used in statistical modeling and inference, in economics, finance, and risk assessment, in quality control and process improvement (lean 6 sigma method), etc. When dealing with data, it's always important to study their distributions before moving to statistical models since the latter are sometimes built on asumptions on the input data. At the era of [generative AI](https://en.wikipedia.org/wiki/Generative_model), probability distributions are a must: what a generative model does is learning the underlying distribution of its training data in order to generate "likely" data points, images, audios, videos, and texts when prompted.
